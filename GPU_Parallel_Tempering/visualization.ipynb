```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
N = 128
REPLICS = 16
lattice = np.fromfile("lattice.bin", dtype=np.int8, count=-1);
energy = np.fromfile("energy.bin", dtype=np.int64, count=-1);
temp = np.fromfile("temp.bin", dtype=np.float64, count=-1);
```

```python
font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 22}
mpl.rc('font', **font)

rows, cols = 4, 4
fig, ax = plt.subplots(rows, cols, figsize=(20, 20)) 
for i in range(rows):
	for j in range(cols):
		offset = j + i * cols
		ax[i, j].imshow(lattice[offset * N * N : (offset+1) * N * N].reshape((N, N)));
		ax[i, j].set_title(f"Temperature: {temp[offset]:.2f}");
		ax[i, j].axis('off')
plt.tight_layout()
```
