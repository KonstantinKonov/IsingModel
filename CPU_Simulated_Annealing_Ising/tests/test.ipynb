```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.fromfile("T.bin", dtype=np.float64, count=-1)
en = np.fromfile("energy.bin", dtype=np.float64, count=-1)
en = en.reshape((31, -1))

en_T_mean = en.mean(axis=1)

en2_T_mean = (en * en).mean(axis=1)

cv = en2_T_mean - en_T_mean * en_T_mean
```
